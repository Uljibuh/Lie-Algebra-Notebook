# First part, understanding Lie Group Axioms
# step by step calculation of each axiom

```python
"""
Step 1: SO(2) Group Axioms — See Every Matrix at Every Step
============================================================
"""

import numpy as np
np.set_printoptions(precision=4, suppress=True)  # clean printing


def R(theta_deg):
    """2D rotation matrix from degrees."""
    t = np.deg2rad(theta_deg)
    c, s = np.cos(t), np.sin(t)
    return np.array([[c, -s],
                     [s,  c]])
 
def show(name, mat):
    """Print a matrix with a label."""
    print(f"  {name} =  [{mat[0,0]:+.4f}  {mat[0,1]:+.4f}]")
    print(f"  {' '*len(name)}    [{mat[1,0]:+.4f}  {mat[1,1]:+.4f}]")
    print()


# ============================================================
# First, let's see what rotation matrices actually look like
# ============================================================
print("="*55)
print("  WHAT DO ROTATION MATRICES LOOK LIKE?")
print("="*55)
print()
 
for angle in [0, 30, 45, 60, 90, 180]:
    show(f"R({angle:>3}°)", R(angle))
```

```python
# ============================================================
# AXIOM 1: CLOSURE — multiply two rotations, get a rotation
# ============================================================
print("="*55)
print("  AXIOM 1: CLOSURE")
print("  R(30°) @ R(45°)  should give  R(75°)")
print("="*55)
print()
 
A = R(30)
B = R(45)
print("  Start with:")
show("R(30°)", A)
show("R(45°)", B)
 
print("  Multiply them:  R(30°) @ R(45°)")
print()
print("  Row 0, Col 0:  (+0.8660)*(+0.7071) + (-0.5000)*( +0.7071)")
print(f"               =  {0.8660*0.7071:.4f}  +  {-0.5000*0.7071:.4f}  =  {0.8660*0.7071 + (-0.5000)*0.7071:.4f}")
print()
print("  Row 0, Col 1:  (+0.8660)*(-0.7071) + (-0.5000)*(+0.7071)")
print(f"               =  {0.8660*(-0.7071):.4f}  +  {-0.5000*0.7071:.4f}  =  {0.8660*(-0.7071) + (-0.5000)*0.7071:.4f}")
print()
print("  (I'll spare you rows 1... here's the full result:)")
print()
 
AB = A @ B
show("R(30°)@R(45°)", AB)
show("R(75°)        ", R(75))
print("  They match! Rotation * Rotation = Rotation. ✓")
print()
 
```

```python
 
# ============================================================
# AXIOM 2: ASSOCIATIVITY
# ============================================================
print("="*55)
print("  AXIOM 2: ASSOCIATIVITY")
print("  (R(30°) @ R(45°)) @ R(60°)  vs  R(30°) @ (R(45°) @ R(60°))")
print("="*55)
print()
 
C = R(60)
 
print("  Path 1: first multiply R(30°) @ R(45°), then @ R(60°)")
print()
step1 = A @ B
show("step1 = R(30°)@R(45°)", step1)
lhs = step1 @ C
show("step1 @ R(60°)       ", lhs)
 
print("  Path 2: first multiply R(45°) @ R(60°), then R(30°) @")
print()
step2 = B @ C
show("step2 = R(45°)@R(60°)", step2)
rhs = A @ step2
show("R(30°) @ step2       ", rhs)
 
print("  Both paths give the same matrix! ✓")
show("(both equal R(135°)) ", R(135))
 
```

```python
# ============================================================
# AXIOM 3: IDENTITY
# ============================================================
print("="*55)
print("  AXIOM 3: IDENTITY")
print("  R(0°) @ R(30°)  should give  R(30°)")
print("="*55)
print()
 
I = R(0)
show("R(0°)  (identity)", I)
show("R(30°)           ", A)
 
print("  Multiply: R(0°) @ R(30°)")
IA = I @ A
show("R(0°) @ R(30°)   ", IA)
print("  Same as R(30°)? Yes! Identity does nothing. ✓")
print()
 
print("  Other direction: R(30°) @ R(0°)")
AI = A @ I
show("R(30°) @ R(0°)   ", AI)
print("  Also same! Works from both sides. ✓")
print()
 
```

```python
# ============================================================
# AXIOM 4: INVERSE
# ============================================================
print("="*55)
print("  AXIOM 4: INVERSE")
print("  R(30°) @ R(-30°)  should give  R(0°) = Identity")
print("="*55)
print()
 
A_inv = R(-30)
show("R(30°)    ", A)
show("R(-30°)   ", A_inv)
 
print("  Notice: R(-30°) is R(30°) with rows/cols swapped!")
print("  That means R(-θ) = R(θ) transposed.")
print()
 
print("  Multiply: R(30°) @ R(-30°)")
product = A @ A_inv
show("R(30°)@R(-30°)", product)
show("R(0°) identity", I)
print("  We get back to identity! ✓")
print()
 
print("="*55)
print("  SUMMARY")
print("="*55)
print("""
  Closure:       R(a) @ R(b) = R(a+b)     — still a rotation
  Associativity: (R@R)@R = R@(R@R)         — grouping doesn't matter
  Identity:      R(0) = I                  — do-nothing rotation
  Inverse:       R(θ)^{-1} = R(-θ) = R(θ)^T  — rotate backwards
 
  SO(2) is a Lie group!  ✓ ✓ ✓ ✓
""")
```

```python
"""
Step 2: Matrix SO(2) vs Complex Number U(1) — Side by Side
============================================================
A 2D rotation R(θ) and the complex number e^{iθ} = cosθ + i sinθ
are TWO REPRESENTATIONS of the SAME group.

Let's see it number by number.
"""
import numpy as np
np.set_printoptions(precision=4, suppress=True)

def R(deg):
    """2D rotation matrix."""
    t = np.deg2rad(deg)
    c, s = np.cos(t), np.sin(t)
    return np.array([[c, -s],
                     [s,  c]])

def Z(deg):
    """Complex number on unit circle: e^{iθ}."""
    return np.exp(1j * np.deg2rad(deg))

def show_both(label, mat, cpx):
    """Print matrix and complex number side by side."""
    print(f"  {'MATRIX':^30s}  |  {'COMPLEX NUMBER':^25s}")
    print(f"  {label}")
    print(f"  [{mat[0,0]:+.4f}  {mat[0,1]:+.4f}]     |  {cpx.real:+.4f} {cpx.imag:+.4f}i")
    print(f"  [{mat[1,0]:+.4f}  {mat[1,1]:+.4f}]     |  |z| = {abs(cpx):.4f}")
    print()

# ============================================================
print("="*60)
print("  WHAT IS THE CONNECTION?")
print("="*60)
print("""
  Matrix:          Complex number:
  [cosθ  -sinθ]    cosθ + i·sinθ  =  e^{iθ}
  [sinθ   cosθ]

  The mapping:
    top-left  = cosθ  →  real part
    bottom-left = sinθ  →  imaginary part
    
  They encode the SAME two numbers (cosθ, sinθ)
  just in different packaging!
""")

# Let's see the actual numbers
print("="*60)
print("  COMPARE: What do they look like at various angles?")
print("="*60)
print()
print(f"  {'ANGLE':^8s} | {'MATRIX [row0]  [row1]':^30s} | {'COMPLEX':^20s}")
print(f"  {'-'*8} | {'-'*30} | {'-'*20}")
for deg in [0, 30, 45, 60, 90, 180]:
    m = R(deg)
    z = Z(deg)
    print(f"  {deg:>5}°  | [{m[0,0]:+.3f} {m[0,1]:+.3f}][{m[1,0]:+.3f} {m[1,1]:+.3f}] | {z.real:+.4f} {z.imag:+.4f}i")
print()
print("  Look: cos and sin appear in BOTH columns!")
print()

# ============================================================
# AXIOM 1: CLOSURE
# ============================================================
print("="*60)
print("  AXIOM 1: CLOSURE")
print("="*60)
print()

print("  --- Matrix: R(30°) @ R(45°) ---")
mA, mB = R(30), R(45)
mAB = mA @ mB
print(f"  R(30°) =  [{mA[0,0]:+.4f}  {mA[0,1]:+.4f}]")
print(f"            [{mA[1,0]:+.4f}  {mA[1,1]:+.4f}]")
print()
print(f"  R(45°) =  [{mB[0,0]:+.4f}  {mB[0,1]:+.4f}]")
print(f"            [{mB[1,0]:+.4f}  {mB[1,1]:+.4f}]")
print()
print(f"  R(30°)@R(45°) = [{mAB[0,0]:+.4f}  {mAB[0,1]:+.4f}]  = R(75°)")
print(f"                  [{mAB[1,0]:+.4f}  {mAB[1,1]:+.4f}]")
print(f"  det = {np.linalg.det(mAB):.4f}  (still a rotation)")
print()

print("  --- Complex: Z(30°) * Z(45°) ---")
zA, zB = Z(30), Z(45)
zAB = zA * zB
print(f"  Z(30°) = {zA.real:+.4f} {zA.imag:+.4f}i")
print(f"  Z(45°) = {zB.real:+.4f} {zB.imag:+.4f}i")
print()
print(f"  Multiply: ({zA.real:.4f} + {zA.imag:.4f}i)({zB.real:.4f} + {zB.imag:.4f}i)")
print(f"  real part: ({zA.real:.4f})({zB.real:.4f}) - ({zA.imag:.4f})({zB.imag:.4f}) = {zAB.real:+.4f}")
print(f"  imag part: ({zA.real:.4f})({zB.imag:.4f}) + ({zA.imag:.4f})({zB.real:.4f}) = {zAB.imag:+.4f}")
print(f"  Z(30°)*Z(45°) = {zAB.real:+.4f} {zAB.imag:+.4f}i")
print(f"  |z| = {abs(zAB):.4f}  (still on unit circle)")
print()

print(f"  Z(75°)         = {Z(75).real:+.4f} {Z(75).imag:+.4f}i")
print()
print("  >>> SAME cos and sin values in both! ✓")
print()

# ============================================================
# KEY: WHY complex multiplication = matrix multiplication
# ============================================================
print("="*60)
print("  WHY ARE THEY THE SAME? (the key insight)")
print("="*60)
print("""
  Complex multiplication (a+bi)(c+di):
      real = ac - bd
      imag = ad + bc

  Matrix multiplication [a -b][c -d]:
                        [b  a][d  c]
      top-left  = ac - bd  = ac + (-b)(d)
      bot-left  = bc + ad  = (b)(c) + (a)(d)

  IDENTICAL arithmetic! The matrix is just the complex
  number written as a 2x2 grid.
""")

# ============================================================
# AXIOM 2: ASSOCIATIVITY
# ============================================================
print("="*60)
print("  AXIOM 2: ASSOCIATIVITY")
print("="*60)
print()

mC = R(60)
zC = Z(60)

print("  --- Matrix path 1: (R(30)@R(45)) @ R(60) ---")
m_lhs = (mA @ mB) @ mC
print(f"  result = [{m_lhs[0,0]:+.4f}  {m_lhs[0,1]:+.4f}]")
print(f"           [{m_lhs[1,0]:+.4f}  {m_lhs[1,1]:+.4f}]")
print()

print("  --- Matrix path 2: R(30) @ (R(45)@R(60)) ---")
m_rhs = mA @ (mB @ mC)
print(f"  result = [{m_rhs[0,0]:+.4f}  {m_rhs[0,1]:+.4f}]")
print(f"           [{m_rhs[1,0]:+.4f}  {m_rhs[1,1]:+.4f}]")
print()

print("  --- Complex path 1: (Z(30)*Z(45)) * Z(60) ---")
z_lhs = (zA * zB) * zC
print(f"  result = {z_lhs.real:+.4f} {z_lhs.imag:+.4f}i")
print()

print("  --- Complex path 2: Z(30) * (Z(45)*Z(60)) ---")
z_rhs = zA * (zB * zC)
print(f"  result = {z_rhs.real:+.4f} {z_rhs.imag:+.4f}i")
print()
print("  All four give: cos(135°) and sin(135°). ✓")
print()

# ============================================================
# AXIOM 3: IDENTITY
# ============================================================
print("="*60)
print("  AXIOM 3: IDENTITY")
print("="*60)
print()

mI = R(0)
zI = Z(0)

print(f"  Matrix identity:   R(0°) = [{mI[0,0]:+.4f}  {mI[0,1]:+.4f}]")
print(f"                            [{mI[1,0]:+.4f}  {mI[1,1]:+.4f}]")
print()
print(f"  Complex identity:  Z(0°) = {zI.real:+.4f} {zI.imag:+.4f}i  =  just 1!")
print()

mIA = mI @ mA
zIA = zI * zA

print(f"  R(0°)@R(30°) = [{mIA[0,0]:+.4f}  {mIA[0,1]:+.4f}]  same as R(30°) ✓")
print(f"                 [{mIA[1,0]:+.4f}  {mIA[1,1]:+.4f}]")
print()
print(f"  Z(0°)*Z(30°) = {zIA.real:+.4f} {zIA.imag:+.4f}i  same as Z(30°) ✓")
print()
print("  Matrix identity = I (2x2)")
print("  Complex identity = 1 (the number one!)")
print()

# ============================================================
# AXIOM 4: INVERSE
# ============================================================
print("="*60)
print("  AXIOM 4: INVERSE")
print("="*60)
print()

mA_inv = R(-30)
zA_inv = Z(-30)

print(f"  R(30°)  = [{mA[0,0]:+.4f}  {mA[0,1]:+.4f}]     Z(30°)  = {zA.real:+.4f} {zA.imag:+.4f}i")
print(f"            [{mA[1,0]:+.4f}  {mA[1,1]:+.4f}]")
print()
print(f"  R(-30°) = [{mA_inv[0,0]:+.4f}  {mA_inv[0,1]:+.4f}]     Z(-30°) = {zA_inv.real:+.4f} {zA_inv.imag:+.4f}i")
print(f"            [{mA_inv[1,0]:+.4f}  {mA_inv[1,1]:+.4f}]")
print()

m_prod = mA @ mA_inv
z_prod = zA * zA_inv

print(f"  R(30°)@R(-30°) = [{m_prod[0,0]:+.4f}  {m_prod[0,1]:+.4f}]  = Identity ✓")
print(f"                   [{m_prod[1,0]:+.4f}  {m_prod[1,1]:+.4f}]")
print()
print(f"  Z(30°)*Z(-30°) = {z_prod.real:+.4f} {z_prod.imag:+.4f}i  = 1 ✓")
print()

print("  Matrix inverse:  R(-θ) = R(θ)^T      (transpose)")
print("  Complex inverse: Z(-θ) = Z(θ)*        (conjugate)")
print("  Transpose ↔ Conjugate:  same operation, different language!")
print()

# ============================================================
# SUMMARY
# ============================================================
print("="*60)
print("  SUMMARY: DICTIONARY BETWEEN THE TWO WORLDS")
print("="*60)
print("""
  MATRIX  SO(2)              COMPLEX  U(1)
  ─────────────────────────────────────────────────
  [cosθ  -sinθ]              cosθ + i·sinθ = e^{iθ}
  [sinθ   cosθ]

  Matrix multiply  @         Complex multiply  *
  Transpose  R^T             Conjugate  z*
  Identity matrix  I         The number  1
  det(R) = 1                 |z| = 1
  Angle addition             Exponent addition:
    R(a)@R(b) = R(a+b)        e^{ia} · e^{ib} = e^{i(a+b)}

  THEY ARE THE SAME GROUP, written differently.
  Mathematicians say: SO(2) ≅ U(1)  (isomorphic)
""")
```

```python
"""
WHY Transpose = Conjugate (for rotation matrices / unit complex numbers)
========================================================================
Let's trace exactly where each number goes.
"""
import numpy as np
np.set_printoptions(precision=4, suppress=True)

deg = 30
t = np.deg2rad(deg)
c, s = np.cos(t), np.sin(t)

print("="*60)
print("  START: R(30°) and Z(30°)")
print("="*60)
print()
print(f"  cos(30°) = {c:+.4f}")
print(f"  sin(30°) = {s:+.4f}")
print()

print("  Matrix R(30°):")
print(f"       col0    col1")
print(f"  row0 [{c:+.4f}  {-s:+.4f}]")
print(f"  row1 [{s:+.4f}  {c:+.4f}]")
print()

print("  Complex Z(30°):")
print(f"  {c:+.4f} + {s:+.4f}i")
print()

# ============================================================
print("="*60)
print("  STEP 1: TRANSPOSE the matrix")
print("="*60)
print()
print("  Transpose = swap row and column indices")
print("  (row0,col1) swaps with (row1,col0)")
print()
print("  BEFORE transpose:           AFTER transpose:")
print(f"  [{c:+.4f}  {-s:+.4f}]          [{c:+.4f}  {s:+.4f}]")
print(f"  [{s:+.4f}  {c:+.4f}]          [{-s:+.4f}  {c:+.4f}]")
print()
print("  What changed?")
print(f"    (row0,col1): {-s:+.4f}  →  {s:+.4f}    sign of sin FLIPPED")
print(f"    (row1,col0): {s:+.4f}  →  {-s:+.4f}    sign of sin FLIPPED")
print(f"    diagonal:    {c:+.4f}  →  {c:+.4f}    cos UNCHANGED")
print()
print("  Transpose flips the sign of sin in the matrix.")
print()

# ============================================================
print("="*60)
print("  STEP 2: CONJUGATE the complex number")
print("="*60)
print()
print("  Conjugate = flip sign of imaginary part")
print()
print(f"  BEFORE conjugate:    AFTER conjugate:")
print(f"  {c:+.4f} + {s:+.4f}i     {c:+.4f} - {s:+.4f}i")
print()
print("  What changed?")
print(f"    real part:  {c:+.4f}  →  {c:+.4f}    cos UNCHANGED")
print(f"    imag part:  {s:+.4f}  →  {-s:+.4f}    sin FLIPPED")
print()
print("  Conjugate flips the sign of sin in the complex number.")
print()

# ============================================================
print("="*60)
print("  STEP 3: SEE THE CONNECTION")
print("="*60)
print()
print("  Matrix layout:")
print(f"  [cos   -sin]    complex number: cos + i·sin")
print(f"  [+sin   cos]")
print()
print("  After transpose:")
print(f"  [cos   +sin]    after conjugate: cos - i·sin")
print(f"  [-sin   cos]")
print()
print("  In the matrix, sin lives in the OFF-DIAGONAL slots.")
print("  Transposing swaps those slots → flips the sign of sin.")
print()
print("  In the complex number, sin lives in the IMAGINARY part.")
print("  Conjugating flips the imaginary part → flips the sign of sin.")
print()
print("  ┌─────────────────────────────────────────────────┐")
print("  │  TRANSPOSE and CONJUGATE both do ONE thing:     │")
print("  │                                                 │")
print("  │      flip the sign of sin(θ)                    │")
print("  │                                                 │")
print("  │  which turns θ into -θ                          │")
print("  │  which REVERSES the rotation                    │")
print("  └─────────────────────────────────────────────────┘")
print()

# ============================================================
print("="*60)
print("  VERIFY WITH NUMBERS")
print("="*60)
print()

R30 = np.array([[c, -s], [s, c]])
R30_T = R30.T
R30_neg = np.array([[np.cos(-t), -np.sin(-t)], [np.sin(-t), np.cos(-t)]])

Z30 = complex(c, s)
Z30_conj = Z30.conjugate()
Z30_neg = complex(np.cos(-t), np.sin(-t))

print("  MATRIX:")
print(f"    R(30°)^T  = [{R30_T[0,0]:+.4f}  {R30_T[0,1]:+.4f}]")
print(f"                [{R30_T[1,0]:+.4f}  {R30_T[1,1]:+.4f}]")
print()
print(f"    R(-30°)   = [{R30_neg[0,0]:+.4f}  {R30_neg[0,1]:+.4f}]")
print(f"                [{R30_neg[1,0]:+.4f}  {R30_neg[1,1]:+.4f}]")
print()
print(f"    Same? {np.allclose(R30_T, R30_neg)}   ← transpose = negate angle")
print()

print("  COMPLEX:")
print(f"    Z(30°)*   = {Z30_conj.real:+.4f} {Z30_conj.imag:+.4f}i")
print(f"    Z(-30°)   = {Z30_neg.real:+.4f} {Z30_neg.imag:+.4f}i")
print()
print(f"    Same? {np.isclose(Z30_conj, Z30_neg)}   ← conjugate = negate angle")
print()

print("  And both give back identity when multiplied:")
print()
prod_m = R30 @ R30_T
prod_z = Z30 * Z30_conj
print(f"    R(30°) @ R(30°)^T = [{prod_m[0,0]:+.4f}  {prod_m[0,1]:+.4f}]  = I ✓")
print(f"                        [{prod_m[1,0]:+.4f}  {prod_m[1,1]:+.4f}]")
print()
print(f"    Z(30°) * Z(30°)*  = {prod_z.real:+.4f} {prod_z.imag:+.4f}i  = 1 ✓")
print()
print(f"    (note: {c:.4f}² + {s:.4f}² = {c**2 + s**2:.4f}  ← cos²+sin²=1 does the work)")
```
